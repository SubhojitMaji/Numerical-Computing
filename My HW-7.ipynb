# Estimating the Number e

The limit 

`e = lim (n → ∞) (1 + 1/n)^n`

defines the number `e` in calculus. To estimate `e`, compute the value of the expression 

`(1 + 1/n)^n`

for `n = 8, 8^2, 8^3, ..., 8^10`. Compare these estimates with the value of `e` obtained from 

`e = exp(1.0)`.

Interpret the results.

````python
import math

def estimate_e(n):
  return (1 + 1/n)**n

def compare_estimates(n_values):
  print("Let's see how well we can estimate 'e' using the limit definition!")
  print("We'll try plugging in different values for 'n' and see what we get.")

  e_exact = math.exp(1.0)  # The 'true' value of e
  print(f"\nThe actual value of 'e' (using exp(1)) is: {e_exact}")

  for n in n_values:
    estimated_e = estimate_e(n)
    error = abs(estimated_e - e_exact) # Calculate the absolute error
    print(f"\nFor n = {n}, our estimate is: {estimated_e}")
    print(f"The difference from the actual value is: {error}")


# Run the comparison with the specified n values
n_values = [8, 8**2, 8**3, 8**4, 8**5, 8**6, 8**7, 8**8, 8**9, 8**10]
compare_estimates(n_values)
````

# Approximating π Using Semicircle Division

The length of the curved part of a unit semicircle is π. We can approximate π by using triangles and elementary mathematics. Consider the semicircle with the arc bisected as in Figure (a). The hypotenuse of the right triangle is √2. Hence, a rough approximation to π is given by 2√2 ≈ 2.8284. In Figure (b), we consider an angle θ that is a fraction 1/k of the semicircle. The secant shown has length 2 sin(θ/2), and so an approximation to π is 2k sin(θ/2). From trigonometry, we have sin²(θ/2) = (1/2)(1 - cos θ) = (1/2)(1 - √(1 - sin² θ)) = sin² θ / (2 + 2√(1 - sin² θ)). Now let θₙ be the angle that results from division of the semicircular arc into 2ⁿ⁻¹ pieces. Next let Sₙ = sin² θₙ and Pₙ = 2ⁿ √(Sₙ₊₁). Show that Sₙ₊₁ = Sₙ / (2 + 2√(1 - Sₙ)) and Pₙ is an approximation to π. Starting with S₂ = 1 and P₁ = 2, compute Sₙ₊₁ and Pₙ recursively for 2 ≤ n ≤ 20.

````python
import math

def approximate_pi(n_max=20):
    s_values = [0] * (n_max + 2)  
    p_values = [0] * (n_max + 1) 

    s_values[2] = 1.0  
    p_values[1] = 2.0  
    print("n\tS_(n+1)\t\tP_n")
    print("------------------------")
    print("1\t-\t\t2.0")  
    for n in range(2, n_max + 1):
        s_values[n + 1] = s_values[n] / (2 + 2 * math.sqrt(1 - s_values[n]))
        p_values[n] = (2 ** n) * math.sqrt(s_values[n + 1])
        print(f"{n}\t{s_values[n+1]:.15f}\t{p_values[n]:15f}")

approximate_pi()
````

integer function nDigit(n, x)

````python
def nDigit(n, x):
    """
    Returns the nth nonzero digit in the decimal expression for the real number x.

    Args:
        n: The index of the nonzero digit to retrieve (1-based).
        x: The real number.

    Returns:
        The nth nonzero digit as an integer, or None if not found.
    """

    if not isinstance(x, (int, float)):
        raise TypeError("x must be a real number")

    if not isinstance(n, int) or n <= 0:
        raise ValueError("n must be a positive integer")

    x_str = str(x)
    
    # Remove the sign if it exists
    if x_str.startswith('-'):
        x_str = x_str[1:]
    
    # Remove leading zeros before the decimal point
    if '.' in x_str:
        integer_part, decimal_part = x_str.split('.')
        integer_part = integer_part.lstrip('0')
        if not integer_part:
            integer_part = '0'
        x_str = integer_part + '.' + decimal_part
    else:
        x_str = x_str.lstrip('0')
        if not x_str:
            x_str = '0'
            
    nonzero_digits = [digit for digit in x_str if digit.isdigit() and digit != '0']

    if n <= len(nonzero_digits):
        return int(nonzero_digits[n - 1])
    else:
        return None  # Or raise an exception, depending on desired behavior

# Example usage
print(nDigit(1, 0.001234)) 
````

# Lebesgue Constants in Fourier Series

In the theory of Fourier series, some numbers known as **Lebesgue constants** play a role. A formula for them is:

\[
\rho_n = \frac{1}{2n+1} + \frac{2}{\pi} \sum_{k=1}^n \frac{1}{k} \tan \frac{\pi k}{2n+1}
\]

Write and run a program to compute \(\rho_1, \rho_2, \ldots, \rho_{100}\) with eight decimal digits of accuracy. Then test the validity of the inequality:

\[
0 \leq \frac{4}{\pi^2} \ln(2n+1) + 1 - \rho_n \leq 0.0106
\]

````python
import math

def compute_rho(n):
    """Compute the Lebesgue constant ρ_n using the given formula."""
    sum_term = 0.0
    for k in range(1, n + 1):
        sum_term += (1 / k) * math.tan((math.pi * k) / (2 * n + 1))
    rho_n = (1 / (2 * n + 1)) + (2 / math.pi) * sum_term
    return rho_n

def test_inequality(rho_n, n):
    """Test the inequality 0 ≤ (4/π²)ln(2n+1) + 1 - ρ_n ≤ 0.0106."""
    left_side = (4 / (math.pi ** 2)) * math.log(2 * n + 1) + 1 - rho_n
    return 0 <= left_side <= 0.0106

# Compute ρ_n for n = 1 to 100
results = []
for n in range(1, 101):
    rho_n = compute_rho(n)
    is_valid = test_inequality(rho_n, n)
    results.append((n, rho_n, is_valid))

# Print results with 8 decimal digits of accuracy
print("n\tρ_n\t\t\tInequality Valid")
for n, rho_n, is_valid in results:
    print(f"{n}\t{rho_n:.8f}\t{is_valid}")
````
