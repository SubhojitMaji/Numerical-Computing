Q8. Write a function `my_fun_plotter(f, x)` where f is a lambda function and x is an array. The
 function should plot f evaluated at x. Remember to label the x- and y-axes.

```python
import numpy as np
import matplotlib.pyplot as plt

def my_fun_plotter(f, x):
    """
    Plot a lambda function f evaluated at points x.
    
    Parameters:
    f (callable): Lambda function to be plotted
    x (array-like): Array of x values where f will be evaluated
    """
    # Evaluate the function at each x point
    y = [f(val) for val in x]
    
    # Create the plot
    plt.figure(figsize=(10, 6))
    plt.plot(x, y, 'b-', linewidth=2)
    
    # Add labels and grid
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.grid(True, linestyle='--', alpha=0.7)
    
    # Display the plot
    plt.show()

# Test the function
x_values = np.linspace(0, 2*np.pi, 100)
test_func = lambda x: np.sqrt(x) + np.exp(np.sin(x))
my_fun_plotter(test_func, x_values)
```

Q9. Write a function with `my_poly_plotter(n,x)` that plots the polynomials $p_k(x) = x^k$ for $k =
 1,...,n$. Make sure your plot has axis labels and a title.

```python
import numpy as np
import matplotlib.pyplot as plt

def my_poly_plotter(n, x):
    """
    Plot polynomials p_k(x) = x^k for k = 1,...,n
    
    Parameters:
    n (int): Maximum degree of polynomials to plot
    x (array-like): Array of x values where polynomials will be evaluated
    """
    plt.figure(figsize=(10, 6))
    
    # Plot each polynomial
    for k in range(1, n+1):
        y = x**k
        plt.plot(x, y, label=f'x^{k}')
    
    # Add labels and title
    plt.xlabel('x')
    plt.ylabel('p_k(x)')
    plt.title(f'Polynomials from degree 1 to {n}')
    
    # Add grid and legend
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.legend()
    
    # Display the plot
    plt.show()

# Test the function
x_values = np.linspace(-2, 2, 200)
my_poly_plotter(4, x_values)
```

Q10. Assume you have three points at the corner of an equilateral triangle, $P_1$= (0,0),$P_2$ = (0.5,
 2/2),
 and $P_3$ =(1,0). Generate another set of points $p_i = (x_i,y_i)$ such that $p_1 = (0,0)$ and $p_i+1$ is the
 midpoint between $p_i$ and $P_1$ with 33% probability, the midpoint between $p_i$ and $P_2$ with 33%  probability, and the midpoint between pi and P3 with 33% probability. Write a function `my_sier
pinski(n)` that generates the points $p_i$ for i = 1,...,n. The function should make a plot of the
 points using blue dots (i.e., "b." as the third argument to the `plt.plot function`).


```python
def my_sierpinski(n):
    """
    Generate and plot points for the Sierpinski triangle.
    
    Parameters:
    n (int): Number of points to generate
    """
    # Define the vertices of the equilateral triangle
    P1 = np.array([0, 0])
    P2 = np.array([0.5, np.sqrt(3)/2])
    P3 = np.array([1, 0])
    
    # Initialize the list of points with the first point
    points = [P1]
    
    # Generate the points
    for _ in range(n - 1):
        current_point = points[-1]
        r = np.random.rand()
        if r < 1/3:
            next_point = (current_point + P1) / 2
        elif r < 2/3:
            next_point = (current_point + P2) / 2
        else:
            next_point = (current_point + P3) / 2
        points.append(next_point)
    
    # Convert the list of points to a numpy array for plotting
    points = np.array(points)
    
    # Plot the points
    plt.figure(figsize=(8, 8))
    plt.plot(points[:, 0], points[:, 1], 'b.', markersize=1)
    plt.title('Sierpinski Triangle')
    plt.axis('equal')
    plt.show()

# Test the function
my_sierpinski(10000)
```

Q11. Assume you are generating a set of points $(x_i,y_i)$ where $x_1$ = 0 and $y_1$ =0. The points $(x_i,y_i)$ for
 i =2,...,n are generated according to the following probabilistic relationship 
 
With 1% probability:  
$ x_i = 0, \quad y_i = 0.16 y_{i-1} $

With 7% probability:  
$ x_i = 0.2x_{i-1} - 0.26y_{i-1}, \quad y_i = 0.23x_{i-1} + 0.22y_{i-1} + 1.6 $

With 7% probability:  
$ x_i = -0.15x_{i-1} + 0.28y_{i-1}, \quad y_i = 0.26x_{i-1} + 0.24y_{i-1} + 0.44 $

With 85% probability:  
$ x_i = 0.85x_{i-1} + 0.04y_{i-1}, \quad y_i = -0.04x_{i-1} + 0.85y_{i-1} + 1.6 $

Write a function `my_fern(n)` that generates the points $ (x_i, y_i) $ for $ i = 1, \dots, n $ and plots them using blue dots. Also use `plt.axis("equal")` and `plt.axis("off")` to make the plot more attractive.


```python
import matplotlib.pyplot as plt
import random

def my_fern(n):
    x, y = 0, 0
    x_points, y_points = [x], [y]

    for _ in range(n):
        r = random.random()
        if r < 0.01:
            x, y = 0, 0.16 * y
        elif r < 0.08:
            x, y = 0.2 * x - 0.26 * y, 0.23 * x + 0.22 * y + 1.6
        elif r < 0.15:
            x, y = -0.15 * x + 0.28 * y, 0.26 * x + 0.24 * y + 0.44
        else:
            x, y = 0.85 * x + 0.04 * y, -0.04 * x + 0.85 * y + 1.6
        
        x_points.append(x)
        y_points.append(y)

    plt.figure(figsize=(8, 12))
    plt.scatter(x_points, y_points, color='blue', s=0.2)
    plt.axis("equal")
    plt.axis("off")
    plt.show()

# Example usage:
my_fern(100000)
```

Q1. Write down the characteristic equation for matrix $A = 
\begin{bmatrix}
3 & 2 \\
5 & 3
\end{bmatrix}$

The characteristic equation for the matrix $ A = \begin{bmatrix} 3 & 2 \\ 5 & 3 \end{bmatrix} $ is:

$ \lambda^2 - 6\lambda - 1 = 0 $

```python
import numpy as np

A = np.array([[3,2],[5,3]])
# Get eigenvalues first
eigenvals = np.linalg.eigvals(A)
# Then get the coefficients of polynomial with these roots
char_poly_coeffs = np.poly(eigenvals)

n = len(char_poly_coeffs)
equation = "^{} ".format(n-1)
for i in range(1, n):
    coef = char_poly_coeffs[i]
    if coef >= 0:
        equation += "+ "
    equation += f"{coef.round(2)}^{n-i-1} "

print("Characteristic Equation:", equation)
```

Q2. Use the above characteristic equation to solve for eigenvalues and eigenvectors of matrix A.

```python
import numpy as np
import scipy.linalg as la

# Assuming char_poly_coeffs and A are defined
eigenvalues = np.roots(char_poly_coeffs)
print("Eigenvalues:", eigenvalues)

eig_vectors = []
I = np.eye(A.shape[0])

for  in eigenvalues:
    A_lambda_I = A -  * I
    eigenvector = la.null_space(A_lambda_I)
    if eigenvector.size == 0:
        print(f"No eigenvector found for  = {}")
    else:
        eig_vectors.append(eigenvector)

print("Eigenvectors:")
for vec in eig_vectors:
    print(vec)
```
